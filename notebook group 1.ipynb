```python
#Content distribution: Movies vs Tvshows
import pandas as pd
import matplotlib.pyplot as plt

# Function to calculate the distribution of content types
def get_content_distribution(df):
    content_distribution = df['type'].value_counts()
    print("\nContent Type Distribution:")
    print(content_distribution)
    return content_distribution

# Function to visualize the distribution of content types
def plot_content_distribution(content_distribution):
    content_distribution.plot(kind='bar', color=['skyblue', 'salmon'])
    plt.title('Content Distribution: Movies vs. TV Shows')
    plt.xlabel('Content Type')
    plt.ylabel('Count')
    plt.show()

# Function to load the dataset
def load_dataset(filePath):
  return pd.read_csv(filePath)

def main():
    filePath = 'netflix_titles.csv'
    df = load_dataset(filePath)
    content_distribution = get_content_distribution(df)
    plot_content_distribution(content_distribution)

if __name__ == "__main__":
    main()
```

```python
#Top 5 countries producing Netflix
import pandas as pd
import matplotlib.pyplot as plt

# Function to get the top N countries producing Netflix content
def get_top_countries(df, n=5):
    top_countries = df['country'].value_counts().head(n)
    print("\nTop Countries Producing Netflix Content:")
    print(top_countries)
    return top_countries

# Function to analyze content by the top countries
def analyze_country_content(df, countries):
    for country in countries.index:
        country_data = df[df['country'] == country]
        print(f"\n--- {country} ---")
        print("Content Ratings Distribution:")
        print(country_data['rating'].value_counts())
        print("\nTop Genres:")
        print(country_data['listed_in'].value_counts().head(5))

# Function to visualize the top countries
def plot_top_countries(countries):
    countries.plot(kind='bar', color='orange')
    plt.title('Top 5 Countries Producing Netflix Content')
    plt.xlabel('Country')
    plt.ylabel('Content Count')
    plt.show()

def main():
    file_path = 'netflix_titles.csv'
    df = pd.read_csv(file_path)
    top_countries = get_top_countries(df)
    analyze_country_content(df, top_countries)
    plot_top_countries(top_countries)

if __name__ == "__main__":
    main()
```

```python
#Top 5 genres by content type
import pandas as pd
import matplotlib.pyplot as plt

def split_data_by_type(df):
    movies = df[df['type'] == 'Movie']
    tv_shows = df[df['type'] == 'TV Show']
    return movies, tv_shows

# Function to find the top genres
def get_top_genres(df, n=5):
    genres = df['listed_in'].str.split(', ', expand=True).stack().value_counts().head(n)
    return genres

# Function to visualize the top genres
def plot_top_genres(movie_genres, tv_genres):
    plt.figure(figsize=(10, 5))
    movie_genres.plot(kind='bar', color='blue', alpha=0.7, label='Movies', width=0.4, position=1)
    tv_genres.plot(kind='bar', color='green', alpha=0.7, label='TV Shows', width=0.4, position=0)
    plt.title('Top Genres by Content Type')
    plt.xlabel('Genres')
    plt.ylabel('Count')
    plt.legend()
    plt.show()

def main():
    file_path = 'netflix_titles.csv'
    df = pd.read_csv(file_path)
    movies, tv_shows = split_data_by_type(df)
    movie_genres = get_top_genres(movies)
    tv_genres = get_top_genres(tv_shows)
    print("\nTop 5 Movie Genres:\n", movie_genres)
    print("\nTop 5 TV Show Genres:\n", tv_genres)
    plot_top_genres(movie_genres, tv_genres)

if __name__ == "__main__":
    main()
```

```python
#Top 10 directors on Netflix
import pandas as pd
import matplotlib.pyplot as plt

# Function to find the top directors
def get_top_directors(df, n=10):
    top_directors = df['director'].value_counts().head(n)
    print("\nTop 10 Directors with Most Content on Netflix:")
    print(top_directors)
    return top_directors

# Function to analyze content type distribution for top directors
def analyze_director_content(df, top_directors):
    print("\nMovies vs. TV Shows for Top Directors:")
    for director in top_directors.index:
        if pd.notna(director):  # Exclude NaN directors
            director_data = df[df['director'] == director]
            print(f"\n--- {director} ---")
            print(director_data['type'].value_counts())

# Function to visualize the top directors
def plot_top_directors(top_directors):
    plt.figure(figsize=(10, 6))
    top_directors.plot(kind='barh', color='teal')
    plt.title('Top 10 Directors on Netflix', fontsize=14)
    plt.xlabel('Count', fontsize=12)
    plt.ylabel('Director', fontsize=12)
    plt.gca().invert_yaxis()  # Invert y-axis to display the highest value on top
    plt.grid(alpha=0.3)
    plt.show()

def main():
    file_path = 'netflix_titles.csv'
    df = pd.read_csv(file_path)
    top_directors = get_top_directors(df)
    analyze_director_content(df, top_directors)
    plot_top_directors(top_directors)

if __name__ == "__main__":
    main()
```

```python
#Netflix content rating distribution
import pandas as pd
import matplotlib.pyplot as plt

# Function to get overall rating distribution
def get_rating_distribution(df):
    rating_distribution = df['rating'].value_counts(dropna=False)
    print("\nOverall Rating Distribution:\n", rating_distribution)
    return rating_distribution

# Function to analyze rating distribution by content type
def analyze_rating_by_type(df):
    rating_by_type = df.groupby('type')['rating'].value_counts(dropna=False)
    print("\nRating Distribution by Content Type:\n", rating_by_type)

# Function to analyze rating distribution for top countries
def analyze_rating_by_country(df, top_n=5):
    top_countries = df['country'].value_counts().head(top_n).index
    for country in top_countries:
        country_data = df[df['country'] == country]
        print(f"\n--- {country} ---")
        print(country_data['rating'].value_counts(dropna=False))

# Function to visualize rating distribution
def plot_rating_distribution(rating_distribution):
    plt.figure(figsize=(10, 6))
    rating_distribution.plot(kind='bar', color='gold')
    plt.title("Netflix Content Ratings Distribution", fontsize=14)
    plt.xlabel("Rating", fontsize=12)
    plt.ylabel("Count", fontsize=12)
    plt.grid(alpha=0.3)
    plt.show()

def main():
    file_path = 'netflix_titles.csv'
    df = pd.read_csv(file_path)
    rating_distribution = get_rating_distribution(df)
    analyze_rating_by_type(df)
    analyze_rating_by_country(df, top_n=5)
    plot_rating_distribution(rating_distribution)

if __name__ == "__main__":
    main()
```

```python
#Top 10 Most Frequently Appearing Actors/Actresses
import pandas as pd
import matplotlib.pyplot as plt

# Function to get top 10 most frequent actors/actresses
def get_top_actors(df):
    cast_series = df['cast'].dropna().str.split(', ').explode()
    top_actors = cast_series.value_counts().head(10)
    print("\nTop 10 Most Frequently Appearing Actors/Actresses:\n", top_actors)
    return top_actors

# Function to visualize the top actors/actresses
def plot_top_actors(top_actors):
    plt.figure(figsize=(10, 6))
    top_actors.plot(kind='barh', color='lightblue')
    plt.title("Top 10 Most Frequently Appearing Actors/Actresses", fontsize=14)
    plt.xlabel("Count", fontsize=12)
    plt.ylabel("Actor/Actress", fontsize=12)
    plt.gca().invert_yaxis()
    plt.grid(alpha=0.3)
    plt.show()

def main():
    file_path = 'netflix_titles.csv'
    df = pd.read_csv(file_path)
    top_actors = get_top_actors(df)
    plot_top_actors(top_actors)

if __name__ == "__main__":
    main()
```

```python
#Top genres for TV-MA content
import pandas as pd
import matplotlib.pyplot as plt

# Function to load and preprocess data
def load_and_preprocess_data(df):
    df['listed_in'] = df['listed_in'].fillna('')  # Handle missing values in 'listed_in'
    genres_series = df['listed_in'].str.split(', ').explode().reset_index(drop=True)
    ratings_series = df['rating'].reset_index(drop=True).repeat(df['listed_in'].str.split(', ').apply(len).fillna(1).astype(int))
    genres_series = genres_series.groupby(level=0).first()
    ratings_series = ratings_series.groupby(level=0).first()
    return genres_series, ratings_series

# Function to create cross-tabulation of genres vs ratings
def create_genre_rating_crosstab(genres_series, ratings_series):
    return pd.crosstab(genres_series, ratings_series)

# Function to get top genres for a specific rating
def get_top_genres_for_rating(crosstab, rating='TV-MA'):
    if rating in crosstab.columns:
        return crosstab[rating].sort_values(ascending=False).head(5)
    else:
        print(f"Rating '{rating}' not found in the dataset.")
        return None

# Function to plot the top genres as a pie chart
def plot_pie_chart(top_genres):
    if top_genres is not None:
        top_genres.plot(kind='pie', autopct='%1.1f%%', colors=['purple', 'blue', 'green', 'red', 'orange'], legend=False)
        plt.title(f"Top Genres for {top_genres.name} Content")
        plt.ylabel('')  # Hide y-axis label
        plt.show()

def main(df):
    genres_series, ratings_series = load_and_preprocess_data(df)
    genres_ratings = create_genre_rating_crosstab(genres_series, ratings_series)
    print("Genres vs Ratings:\n", genres_ratings)
    top_tv_ma_genres = get_top_genres_for_rating(genres_ratings, rating='TV-MA')
    if top_tv_ma_genres is not None:
        print("\nTop 5 Genres for TV-MA:\n", top_tv_ma_genres)
    plot_pie_chart(top_tv_ma_genres)

if __name__ == "__main__":
    df = pd.read_csv('netflix_titles.csv')
    main(df)
```

```python
#Content addition by month
import pandas as pd
import matplotlib.pyplot as plt

# Function to load and preprocess the data
def load_data(file_path):
    df = pd.read_csv(file_path)
    return df

# Function to extract the month from 'date_added' column
def extract_month(df):
    df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')  # Convert to datetime, handle errors
    df['month_added'] = df['date_added'].dt.month  # Extract the month
    return df

# Function to count content additions by month
def count_additions_by_month(df):
    content_by_month = df['month_added'].value_counts().sort_index()  # Count and sort by month
    return content_by_month

# Function to plot content additions by month
def plot_content_by_month(content_by_month):
    content_by_month.plot(kind='bar', color='orange')
    plt.title("Content Additions by Month")
    plt.xlabel("Month")
    plt.ylabel("Number of Additions")
    plt.show()

def main(file_path):
    df = load_data(file_path)
    df = extract_month(df)  # Extract month from 'date_added'
    content_by_month = count_additions_by_month(df)  # Count additions by month
    print("Content Additions by Month:\n", content_by_month)
    plot_content_by_month(content_by_month)

if __name__ == "__main__":
    file_path = 'netflix_titles.csv'
    main(file_path)
```

```python
#Content Additions by Year
import pandas as pd
import matplotlib.pyplot as plt

# Function to load and preprocess the data
def load_data(file_path):
    df = pd.read_csv(file_path)
    return df

# Function to extract the year from 'date_added'
def extract_year(df):
    df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')  # Convert to datetime, handle errors
    df['year_added'] = df['date_added'].dt.year  # Extract the year
    return df

# Function to count content additions by year
def count_additions_by_year(df):
    content_by_year = df['year_added'].value_counts().sort_index()  # Count and sort by year
    return content_by_year

# Function to plot content additions by year
def plot_content_by_year(content_by_year):
    content_by_year.plot(kind='line', marker='o', color='purple')
    plt.title("Netflix Content Additions Over Time")
    plt.xlabel("Year")
    plt.ylabel("Number of Additions")
    plt.grid(True)
    plt.show()

def main(file_path):
    df = load_data(file_path)
    df = extract_year(df)
    content_by_year = count_additions_by_year(df)
    print("Content Additions by Year:\n", content_by_year)
    plot_content_by_year(content_by_year)

if __name__ == "__main__":
    file_path = 'netflix_titles.csv'
    main(file_path)
```

```python
#Movies run time over
import pandas as pd
import matplotlib.pyplot as plt

# Function to load the data
def load_data(file_path):
    df = pd.read_csv(file_path)
    return df

# Function to preprocess the data: Extract runtime for movies and seasons for TV shows
def preprocess_data(df):
    movies = df[df['type'] == 'Movie'].copy()
    tv_shows = df[df['type'] == 'TV Show'].copy()
    movies.loc[:, 'runtime'] = pd.to_numeric(movies['duration'].str.replace(' min', ''), errors='coerce')
    tv_shows.loc[:, 'seasons'] = pd.to_numeric(tv_shows['duration'].str.replace(' Season', '').str.replace('s', ''), errors='coerce')
    return movies, tv_shows

# Function to calculate and print average runtime and seasons
def calculate_averages(movies, tv_shows):
    average_runtime = movies['runtime'].mean()
    average_seasons = tv_shows['seasons'].mean()
    print(f"Average Runtime for Movies: {average_runtime:.2f} minutes")
    print(f"Average Number of Seasons for TV Shows: {average_seasons:.2f}")

# Function to create scatterplot for movie runtime over time
def plot_movie_runtime(movies):
    plt.figure(figsize=(10, 6))
    plt.scatter(movies['release_year'], movies['runtime'], alpha=0.5, color='blue')
    plt.title("Movies' Runtime Over Time", fontsize=14)
    plt.xlabel("Release Year", fontsize=12)
    plt.ylabel("Runtime (Minutes)", fontsize=12)
    plt.grid(alpha=0.3)
    plt.show()

def main(file_path):
    df = load_data(file_path)
    movies, tv_shows = preprocess_data(df)
    calculate_averages(movies, tv_shows)
    plot_movie_runtime(movies)


if __name__ == "__main__":
    file_path = 'netflix_titles.csv'
    main(file_path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to load and preprocess data
def load_and_prepare_data(file_path):
    df = pd.read_csv(file_path)

    # Filter out rows where 'country' is 'Unknown' and reset index
    region_ratings_data = df[df['country'] != 'Unknown'].copy()
    region_ratings_data = region_ratings_data.reset_index(drop=True)
    region_ratings_exploded = region_ratings_data.assign(
        country=region_ratings_data['country'].str.split(',')
    ).explode('country')
    region_ratings_exploded['country'] = region_ratings_exploded['country'].str.strip()
    return region_ratings_exploded

# Function to compute the rating distribution by region
def compute_rating_distribution(region_data):
    rating_by_region = region_data.groupby(['country', 'rating']).size().unstack(fill_value=0)
    return rating_by_region

# Function to visualize the data
def visualize_rating_distribution(rating_by_region):
    top_regions = rating_by_region.loc[rating_by_region.sum(axis=1).nlargest(10).index]
    plt.figure(figsize=(14, 8))
    plt.imshow(top_regions, cmap="coolwarm", aspect='auto')
    plt.colorbar(label="Frequency")
    plt.xticks(range(top_regions.columns.size), top_regions.columns, rotation=45, ha='right')
    plt.yticks(range(top_regions.index.size), top_regions.index)
    plt.title("Frequency of Ratings Across Top Regions")
    plt.xlabel("Ratings")
    plt.ylabel("Regions")

    for i in range(top_regions.shape[0]):
        for j in range(top_regions.shape[1]):
            plt.text(j, i, top_regions.iloc[i, j], ha='center', va='center', color='black')

    plt.tight_layout()
    plt.show()

def main():
    file_path = 'netflix_titles.csv'
    region_data = load_and_prepare_data(file_path)
    rating_by_region = compute_rating_distribution(region_data)
    visualize_rating_distribution(rating_by_region)

    top_regions = rating_by_region.sum(axis=1).nlargest(10)
    print("Top regions with the highest frequency of ratings:\n", top_regions)

if __name__ == "__main__":
    main()
```
